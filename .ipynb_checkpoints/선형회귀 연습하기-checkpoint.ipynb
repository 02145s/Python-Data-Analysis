```python
# import libraries
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 샘플데이터, 결과는 뭘로 예상됩니까?
A = np.array([1,2,3,4,5])
B = np.array([1,2,3,4,5])
```

```python
#선형회귀 결과 linregress=linear-regression=선형회귀
slope, intercept, r_value, p_value, std_err = stats.linregress(A,B)
result = stats.linregress(A,B)
```

```python
# 시각화
plt.plot(A, B, 'o', label='original data')
plt.plot(A, intercept + slope*A, 'r', label='fitted line')
plt.legend()
plt.show()
print(result)
# slope기울기 intercept=y절편 rvalue=Pearson -r
```

```python
# 샘플데이터, 결과는 뭘로 예상됩니까?
A = np.array([1,2,3,4,5])
B = np.array([2,4,6,8,10])
```

```python
#선형회귀 결과 linregress=linear-regression=선형회귀
slope, intercept, r_value, p_value, std_err = stats.linregress(A,B)
result = stats.linregress(A,B)
```

```python
# 시각화
plt.plot(A, B, 'o', label='original data')
plt.plot(A, intercept + slope*A, 'r', label='fitted line')
plt.legend()
plt.show()
print(result)
# slope기울기 intercept=y절편 rvalue=Pearson -r
```

```python
#위에 결과로 알 수 있는 것은 기울기가 2로 변경, y절편 0, y=2x그래프가 잘 찾아진것을 확인할 수 있다. 데이터만 바꿔서 넣으면 알아서 찾아준다.
```
